```python

with open ('balancesheet_sp500.txt','r') as input:
        balancesheet = ast.literal_eval(input.read())
with open ('incomestatement_sp500.txt' , 'r' ) as input : 
        incomestatement = ast.literal_eval(input.read()) 
with open('cashflowstatement_sp500.txt' , 'r' ) as input: 
        cashflowstatement = ast.literal_eval(input.read())

    

  

```

```python
def StockList():
    sp500list = [] 
    requrl = "https://www.slickcharts.com/sp500"
    header = {
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36',
    }

    sp500 = [requrl]
    for i, val in enumerate(sp500):
        res = requests.get(requrl, headers=header)
        webscrape = BeautifulSoup(res.text, 'html.parser')
        listscrape = webscrape.findAll(
            'table',  class_='table table-hover table-borderless table-sm')[0].findAll('tbody')
        for index, value in enumerate(listscrape[0]):
            if len(value) > 1:
                text = re.sub(r"[<td>]", "", str(value))
                text1 = re.split('/', text)

                for i in range(len(text1[3])):
                    if text1[3][i] == '"':
                        sp500list.append((text1[3][0:i]))

    stockset = set(sp500list)
    if len(stockset) != len(sp500list):
        print("Contains Duplicates")
    else:
        print("No Duplicates")

    return sp500list
    
sp500list = StockList()

 


```

```python
faulty = set() 
class Financialstatements():
    def __init__(self, frequency='Annual'):
        self.yearlydict = {'2021':{} , '2020' : {} , '2019' : {} , '2018' : {} } 
        self.frequency = frequency
        

class BalanceSheets(Financialstatements):
    def __init__(self, *args, **kwargs):
        self.yearlydict = {'2021':{} , '2020' : {} , '2019' : {} , '2018' : {} } 
        self.dfB = pd.DataFrame(self.yearlydict)
        super().__init__(*args, **kwargs)



        
class IncomeStatements(Financialstatements):
    def __init__(self,*args, **kwargs):
        self.yearlydict = {'2021':{} , '2020' : {} , '2019' : {} , '2018' : {} } 
        self.dfI = pd.DataFrame(self.yearlydict)
        super(*args, **kwargs)




class CashflowStatements(Financialstatements):
    def __init__(self,*args, **kwargs):
        self.yearlydict = {'2021':{} , '2020' : {} , '2019' : {} , '2018' : {} } 
        self.dfC = pd.DataFrame(self.yearlydict)
        super(*args, **kwargs)





class Stock():
    def __init__(self, ticker, freq='Annual'):
        self.ticker = ticker
        self.balance_sheet = BalanceSheets()
        self.income_statement = IncomeStatements()
        self.cashflow_statement = CashflowStatements()
        
        
    def allocate(self):
            ticker = self.ticker 
            yearlydicB = self.balance_sheet.yearlydict
            yearlydicI = self.income_statement.yearlydict 
            yearlydicC = self.cashflow_statement.yearlydict 

            try : 

                for yearlystatements in balancesheet[ticker]: 
                    for year,statements in yearlystatements.items(): 
                        for year1 in yearlydicB.keys() : 
                            if year[0:4] == year1 : 
                                yearlydicB[year1] = statements 

                for yearlystatements in incomestatement[ticker]: 
                    for year,statements in yearlystatements.items(): 
                        for year1 in yearlydicI.keys() : 
                            if year[0:4] == year1 : 
                                yearlydicI[year1] = statements

                for yearlystatements in cashflowstatement[ticker]: 
                    for year,statements in yearlystatements.items(): 
                        for year1 in yearlydicC.keys() : 
                            if year[0:4] == year1 : 
                                yearlydicC[year1] = statements
            except : (faulty.add(self.ticker))
           
            
    
 
       
        
```
